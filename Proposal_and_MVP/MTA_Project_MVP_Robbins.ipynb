## Analysis of MTA Traffic and Vaccination Rates across New York City
By Samuel Robbins

The goal of this project is to inform the placement of pop-up vaccination clinics in the lowest vaccinated areas of NYC based on the volume of traffic through the neighborhoods subway stations. As Delta continues to spread and the city attempts to reach the remainder of unvaccinated residents, city officials want to put clinics in the lowest vaccinated neighborhoods in areas that see high foot traffic - the subways.

To start exploring the best places to set up vaccination clinics, I looked at the busiest - defined as the sum of entries and exits - stations for the months of June and July.

I then used horizontal barplots to isolate the 20 zipcodes with the highest and lowest vaccination rates. 

From this initial analysis it is clear that there are disparities in the rate of vaccination between neighborhoods, with the highest vaccination rates >80% and the lowest <40%. As I continue this analysis, I will find the busiest stations within the lowest vaccinated neighborhoods.
